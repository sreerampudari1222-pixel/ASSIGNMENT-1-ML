```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate synthetic dataset
np.random.seed(42)  # reproducibility
n_samples = 200
X = np.random.uniform(0, 5, n_samples)
epsilon = np.random.normal(0, 1, n_samples)  # Gaussian noise
y = 3 + 4 * X + epsilon

# Add intercept term for regression
X_b = np.c_[np.ones((n_samples, 1)), X]

# Plot raw data
plt.scatter(X, y, alpha=0.6, label="Data")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Raw Data")
plt.legend()
plt.show()

# 2. Closed-form solution (Normal Equation)
theta_closed = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
intercept_closed, slope_closed = theta_closed
print("Closed-form solution:")
print(f"Intercept: {intercept_closed:.4f}, Slope: {slope_closed:.4f}")

# Plot fitted line
plt.scatter(X, y, alpha=0.6, label="Data")
plt.plot(X, intercept_closed + slope_closed * X, color="red", label="Closed-form fit")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Closed-form Solution")
plt.legend()
plt.show()

# 3. Gradient Descent implementation
def compute_mse(X_b, y, theta):
    predictions = X_b @ theta
    return np.mean((predictions - y) ** 2)

eta = 0.05  # learning rate
n_iterations = 1000
theta = np.zeros(2)  # [intercept, slope]
loss_history = []

for i in range(n_iterations):
    gradients = 2 / n_samples * X_b.T @ (X_b @ theta - y)
    theta -= eta * gradients
    loss_history.append(compute_mse(X_b, y, theta))

intercept_gd, slope_gd = theta
print("\nGradient Descent solution:")
print(f"Intercept: {intercept_gd:.4f}, Slope: {slope_gd:.4f}")

# Plot loss curve
plt.plot(range(n_iterations), loss_history, label="MSE")
plt.xlabel("Iterations")
plt.ylabel("MSE Loss")
plt.title("Loss Curve (Gradient Descent)")
plt.legend()
plt.show()

# Plot GD fitted line
plt.scatter(X, y, alpha=0.6, label="Data")
plt.plot(X, intercept_gd + slope_gd * X, color="green", label="Gradient Descent fit")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent Solution")
plt.legend()
plt.show()

# 4. Comparison
print("\n--- Comparison ---")
print(f"Closed-form -> Intercept: {intercept_closed:.4f}, Slope: {slope_closed:.4f}")
print(f"Gradient Descent -> Intercept: {intercept_gd:.4f}, Slope: {slope_gd:.4f}")
print("\nComment: The Gradient Descent solution should converge close to the Closed-form solution, "
      "showing both methods give nearly identical estimates for intercept and slope.")
```
